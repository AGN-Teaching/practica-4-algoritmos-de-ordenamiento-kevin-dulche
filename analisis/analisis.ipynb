```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("./csv/10000_tiempos.csv")
df
```

```python
InsertionS = df['Insertion Sort']
SelectionS = df['Selection Sort']
BubbleSort = df['Bubble Sort']
MergeSort = df['Merge Sort']
Quicksort = df['Quicksort']
```

```python
# Calcular la desviación estándar para cada algoritmo
std_insertion_sort = df['Insertion Sort'].std()
std_selection_sort = df['Selection Sort'].std()
std_bubble_sort = df['Bubble Sort'].std()
std_merge_sort = df['Merge Sort'].std()
std_quicksort = df['Quicksort'].std()

# Calcular la desviación estándar para cada algoritmo
mean_insertion_sort = df['Insertion Sort'].mean()
mean_selection_sort = df['Selection Sort'].mean()
mean_bubble_sort = df['Bubble Sort'].mean()
mean_merge_sort = df['Merge Sort'].mean()
mean_quicksort = df['Quicksort'].mean()

print("Desviación Estandar")
print(std_insertion_sort)
print(std_selection_sort)
print(std_bubble_sort)
print(std_merge_sort)
print(std_quicksort)

print()

print("Promedio")
print(mean_insertion_sort)
print(mean_selection_sort)
print(mean_bubble_sort)
print(mean_merge_sort)
print(mean_quicksort)
```

```python
# Crear un DataFrame para la desviación estándar de cada algoritmo
std_df = pd.DataFrame({
    'Algoritmo': ['Insertion Sort', 'Selection Sort', 'Bubble Sort', 'Merge Sort', 'Quicksort'],
    'Desviación Estándar': [std_insertion_sort, std_selection_sort, std_bubble_sort, std_merge_sort, std_quicksort]
})
```

```python
# Graficar la desviación estándar
plt.figure(figsize=(10, 6))
plt.bar(std_df['Algoritmo'], std_df['Desviación Estándar'], color='skyblue')
plt.title('Desviación Estándar de Tiempos de Ejecución por Algoritmo')
plt.xlabel('Algoritmo')
plt.ylabel('Desviación Estándar')
plt.show()
```

```python
# Crear gráficos de dispersión para cada algoritmo con la línea de desviación estándar
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Lista de algoritmos
algoritmos = ['Insertion Sort', 'Selection Sort', 'Bubble Sort', 'Merge Sort', 'Quicksort']

for i, algoritmo in enumerate(algoritmos):
    row = i // 3
    col = i % 3
    ax = axs[row, col]
    ax.scatter(df['Ejecución'], df[algoritmo])
    ax.axhline(df[algoritmo].mean(), color='red', linestyle='dashed', linewidth=2, label='Media')
    ax.axhline(df[algoritmo].mean() + df[algoritmo].std(), color='orange', linestyle='dashed', linewidth=2, label='+1 Desviación Estándar')
    ax.axhline(df[algoritmo].mean() - df[algoritmo].std(), color='orange', linestyle='dashed', linewidth=2, label='-1 Desviación Estándar')
    ax.set_title(f'{algoritmo}\nDesviación Estándar: {df[algoritmo].std():.2e}\nPromedio: {df[algoritmo].mean():.2e}')
    ax.set_xlabel('Ejecución')
    ax.set_ylabel('Tiempo de Ejecución (segundos)')
    ax.legend()

axs[1,2].axis("off")

# Añadir títulos a los ejes
fig.text(0.5, 0.04, 'Ejecución', ha='center', va='center', fontsize=14)
fig.text(0.06, 0.5, 'Tiempo de Ejecución', ha='center', va='center', rotation='vertical', fontsize=14)

# Ajustar márgenes
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)

plt.show()
```
